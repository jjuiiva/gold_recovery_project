### Проверим, что эффективность обогащения рассчитана правильно

Нам нужно вручную посчитать значение колонки rougher.output.recovery и сравнить наши вычисления с данными из обучающего датасета.

**Формула:**

Recovery = C * (F - T) / F * (C - T) * 100% , где:
- C - доля золота в концентрате после флотации (rougher.output.concentrate_au);
- F - доля золота в сырье до флотации (rougher.input.feed_au); 
- T - доля золота в отвальных хвостах после флотации (rougher.output.tail_au); 

```python
# создаем отдельные переменные с соответствующими колонками
C = train_df['rougher.output.concentrate_au']
F = train_df['rougher.input.feed_au']
T = train_df['rougher.output.tail_au']
```

```python
# наша посчитанная эффективность
recovery = (C * (F - T)) / (F * (C - T)) * 100
recovery
```

Посчитанная эффективность на предприятии находится в колонке rougher.output.recovery. 
Теперь нам нужно посчитать среднюю абсолютную ошибку среди этих двух переменных.  

Если она будет незначительная, то на этих данных можно обучать модель для предсказаний; если нет - то пишем заказчику и сообщаем об ошибке в данных.

```python
train_df['rougher.output.recovery'].head()
```

Считаем MAE:

```python
# считаем MAE

print('Средняя абсолютная ошибка: {:.2f}'.format(mean_absolute_error(recovery, train_df['rougher.output.recovery'])))
print('Среднее нашей посчитанной эффективности: {:.2f}'.format(recovery.mean()))
print('Среднее эффективности из данных: {:.2f}'.format(train_df['rougher.output.recovery'].mean()))
```

Все данные совпадают. Значит, мы их можем использовать в дальнейшей работе.

### Проанализируем признаки, недоступные в тестовой выборке.

Видим, что у обучающей и тестовой выборки разное кол-во столбцов. Посчитаем, сколько столбцов у них совпадает.

```python
# создадим функцию, которая на выходе будет выдавать списки совпадающих и несовпадающих колонок
def similar_columns (first_df, second_df):
    
    first_columns = first_df.columns
    second_columns = second_df.columns
    similar_columns = []
    other_columns = []
    
    for i in range(len(first_columns)):
        
        if first_columns[i] in second_columns:
            similar_columns.append(first_columns[i])
        else:
            other_columns.append(first_columns[i])
            
    return(similar_columns, other_columns)
```

Можем видеть, что некоторые колонки между собой коррелируют с достаточно высоким показателем (близко к 1 или -1). Это может затруднить обучение модели. 

Напишем функцию, которая будет:
- Для начала исключит колонку date из списка колонок, которые нужно будет удалить;  
- Считать в цикле коррекляцию всех колонок по очереди;
- Если корреляция между двумя колонками будет равна от -1 до -0,8 и от 0,8 до 1, то будем ее считать слишком высокой;
- Среди двух колонок с высокой корреляцией оставим только одну: ту, у которой корреляция с целеым признаком будет выше;
- На выходе выдаст список с названиями колонок, которые нужно будет удалить из датафрейма;

```python
# создаем функцию

def delete_high_corr (df):
    
    columns = df.drop('date', axis=1).columns
    columns_to_delete = []
    
    for i in range(len(columns)):
        
        first_column = columns[i]
        
        for j in range(len(columns)):
            
            second_column = columns[j]
            
            if first_column != second_column:
                
                if (df[first_column].corr(df[second_column]) > 0.8) or (df[first_column].corr(df[second_column]) < -0.8):
                    
                    if df[first_column].corr(y_train_rougher) > df[second_column].corr(y_train_rougher):
                        columns_to_delete.append(second_column)
                    else:
                        columns_to_delete.append(first_column)
                        
    return pd.Series(columns_to_delete).unique()  
```

```python
columns = delete_high_corr(train_df)
columns
```

Функция работает теперь применим ее к нашему обучающему датасету и выведем хитмеп снова:

## Модель

### Напишем функцию для вычисления итоговой метрики качества.

SMAPE - симметричное среднее абсолютное процентное отклонение. Она одинаково учитывает масштаб и целевого признака, и предсказания модели.

**Формула:** 

где:  
- n - кол-во объектов в выборке;  
- Ft - значение фактического целевого признака для объекта под номером t;  
- At - значение предсказания признака для объекта под номером t;  

Создадим функцию для рассчета SMAPE.

```python
def smape (y, y_pred):
    
    smp = 0 
    smp = (abs(y - y_pred) / ((abs(y) + abs(y_pred)) / 2)).sum()
    smp = (100 / len(y)) * smp
    
    return smp
```

Теперь сделаем свой скорринг для оценки наших моделей в GridSearchCV. Используем для этого функцию make_scorer.

```python
SMAPE = make_scorer(smape, greater_is_better=False)
SMAPE
```

Создадим фукцию для рассчета итогового SMAPE:

```python
def total_smape (first_smape, second_smape):
     return (0.25 * first_smape + 0.75 * second_smape)
```
